```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df_countries = pd.get_dummies(df['country'])
df_countries

countries = list(df_countries.columns)

df = pd.concat([df, df_countries], axis=1) #combine one hot encoded country features + original
df.drop(columns = ['country'], inplace = True)
df
```

<b>One Hot Encoding:</b> First, we took the dataframe we got from the world happiness report and noticed that all of the features were numerical except one, which is country. Since country is a categorical feature, we one hot encode it and reassign our dataframe to have the country column replaced by our one hot encoded columns.

<b>Set Definitions</b>: We then defined our X and y variables, or our label and feature variables. Based on what we concluded ourselves about what kind of features would have a high correlation with our label. We did this by reading the World Happiness Report and noting which features could be influential to our label.

```python
features = ['year','Generosity', 'Freedom to make life choices', 'Democratic Quality', 'Log GDP per capita', 'Social support', 'Perceptions of corruption', 'Negative affect', 'Delivery Quality']
y = df['Positive affect']
X = df[features+countries] #using one-hot encoded countries + original features to compile total feature list considered
X.describe()
```

<b>Filling in NA Values with the Mean</b>: For each column, we first checked the sum of the NA values in every column and then filled in missing values with the mean of that column.

```python
#replacing NaN values
for col in X.columns:
    df[col] = df[col].fillna(df[col].mean()) #replacing NaN values
df['Positive affect'] = df['Positive affect'].fillna(df['Positive affect'].mean())

#redifining X and y
all_features = features+countries
X = df[all_features]
y = df['Positive affect']

#print missing to check
missing = X.isna()
print(missing.sum())
missing = y.isna()
print('Positive affect missing: ' + str(missing.sum()))
```

<b>Training the model</b>: We split the dataset into training and testing sets. We also decided to have the test size be 20%. Then we fit the model and produced predictions.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train) #fitting
prediction = model.predict(X_test)
```

```python
#Model Summary for comparison
print('Model Summary:\n')

# Print intercept (alpha)
print('Intercept:')
print('alpha = ' , model.intercept_)

# Print weights
print('\nWeights:')
i = 0
for w in model.coef_:
    print('w_',i+1,'= ', w, ' [ weight of ', all_features[i],']')
    i += 1
```

<b>Validation</b>: We measure validation by looking at the RMSE, R^2 and cross_val_score that gives us the scores for cross validation. 

```python
#RSME, R^2, and CV for validation printing
print('Model Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
print('R^2 =   %.2f'
      % r2_score(y_test, prediction))
print('Cross Validation Mean Squared Error Scores =   '+ str(cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')))
```

<b>Evaluation</b>: We evaluate the model overall by using the model.score function.

```python
print('Model Accuracy\n\naccuracy = ')
print(model.score(X_test, y_test))
```

<b>Feature Selection</b>: We perform feature selection by making a correlation matrix out of our selected features. Then, we look at the correlation specifically for our label, positive affect, and sort the correlation values so that we can pick the top 100 that are the most correlated with our label. 

```python
special_df = df[features+countries+['Positive affect',]] #Redefining our features + labels dataframe
corr_matrix = round(special_df.corr(),5) #creating correlation matrix
corr_matrix
```

```python
corrs = corr_matrix['Positive affect'] #correlations for label
corrs
```

```python
corrs_sorted = corrs.sort_values(ascending=False)
corrs_sorted
```

```python
top_hund_corr = corrs_sorted[1:100].index #isolating top 100 correlated values for label
top_hund_corr
```

```python
df_corrs = df[list(top_hund_corr)+['Positive affect',]] #adding label back to view
df_corrs
print(df_corrs.columns)
```

<b>Training and Testing the model again </b>: With our top 100 correlated features, we retrain the model with the same test_size and model.

```python
#Reassigning X and y
y = df_corrs['Positive affect']
df_corrs.drop(columns ='Positive affect', inplace=True)
X = df_corrs
print(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42) #Splitting data again

#Define model again
model = LinearRegression()
model.fit(X_train, y_train)
prediction = model.predict(X_test)
```

<b>Validation again</b>: We perform validation again on the top 100 correlated values, and see that the validation scores performed worse.

```python
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

<b>Evaluation again</b>: We evaluate the new model on the top 100 correlated values. 

```python
print('Model Accuracy\n\naccuracy = ')
print(model.score(X_test, y_test))
```

<b>Results</b>: Initially, we decided on a set of features based on our opinions of what would be the most correlated with our label, which resulted in the final accuracy of 0.82. We then did feature selection to try to narrow down our feature list to the top 100 most correlated features to see if we could get a higher accuracy. However, the accuracy for the reduced feature list was 0.78, which is lower than the original score. We believe this is due to the fact that our original set of features provided a more informative dataset for the model to make predictions from. If we had reduced the dataset and found that the accuracies were higher, this would indicate that our original model was overfitting, but because our accuracy only decreased after the feature reduction, this means that our original model had minimal overfitting.

Our validation scores for R^2, RSME, and Cross-Validation are very similar to the accuracy score, since for our LinearRegression, we did not alter any hyperparameters, as none of the hyperparameters were applicable to our problem and model.
